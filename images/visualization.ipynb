# This code will visualize the out put of our vlsi model optimization


```python
import matplotlib.pyplot as plt

def visualize_circuits(width, height, x_dims, y_dims, x_vars, y_vars):
    fig, ax = plt.subplots(figsize=(width, height))
    ax.set_xlim(0, width)
    ax.set_ylim(0, height)
    ax.set_aspect('equal')

    colors = ['red', 'black', 'green', 'orange', 'white', 'pink', 'brown', 'gray', 'yellow', 'olive',"magneta","purple"]
    for i in range(len(x_dims)):
        x_pos = x_vars[i]
        y_pos = y_vars[i]
        rect = plt.Rectangle((x_pos, y_pos), x_dims[i], y_dims[i], color=colors[i % len(colors)])
        ax.add_patch(rect)
    plt.show()

width = 13
height = 13
x_dims = [3,3,3,3,3,3,4,4,7]
y_dims = [3,4,5,6,7,8,3,4,6]

x_vars =[7, 7, 10, 0, 0, 10, 3, 3, 3]
y_vars = [10, 0, 0, 0, 6, 5, 10, 0, 4]

visualize_circuits(width, height, x_dims, y_dims, x_vars, y_vars)
```
